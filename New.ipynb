```python
import pandas as pd
import numpy as np
```

```python
nameofcol = ['AgencyNumber','AgencyName','CardholderLastName','CardholderFirstInitial','Description','Amount','Vendor','TransactionDate','PostedDate','MerchantCategoryCode(MCC)']
```

```python
df2010 = pd.read_csv('PCard_Data FY2010.csv', encoding='utf-8',names = nameofcol, skiprows =1,dtype=str)
df2011 = pd.read_csv('PCard_Data FY2011.csv', encoding='utf-8',names = nameofcol, skiprows =1,dtype=str)
df2012 = pd.read_csv('PCard_Data FY2012.csv', encoding='utf-8',names = nameofcol, skiprows =1,dtype=str)
df2013 = pd.read_csv('PCard_Data FY2013.csv', encoding='utf-8',names = nameofcol, skiprows =1,dtype=str)
df2014 = pd.read_csv('PCard_Data FY2014.csv', encoding='utf-8',names = nameofcol, skiprows =1,dtype=str)
df2015 = pd.read_csv('PCard_Data FY2015.csv', encoding='utf-8',names = nameofcol, skiprows =1,dtype=str)
df2015.sample(5)
```

```python
combined = pd.concat([df2010,df2011,df2012,df2013,df2014,df2015])
combined.info()
```

```python
combined['Amount']= combined['Amount'].str.replace('$','')
combined['Amount']= combined['Amount'].str.replace('(', '-')
combined['Amount']= combined['Amount'].str.replace(')', '')
combined['Amount']= combined['Amount'].str.replace(',', '')
combined['Amount']= combined['Amount'].str.strip()
combined.CardholderLastName = combined.CardholderLastName.str.upper()
```

```python
combined['Description']=combined['Description'].str.strip()
combined['Vendor']=combined['Vendor'].str.strip()
combined['MerchantCategoryCode(MCC)']=combined['MerchantCategoryCode(MCC)'].str.strip()
```

```python
combined.TransactionDate = combined.TransactionDate.str.replace('00:00:00','')
combined.TransactionDate = combined.TransactionDate.str.replace('0:00','')
combined.sample(8)
```

```python
combined.PostedDate = combined.PostedDate.str.replace('00:00:00','')
combined.PostedDate = combined.PostedDate.str.replace('0:00','')
```

```python
combined = combined.dropna(subset=['TransactionDate','PostedDate']).copy()
```

```python
dfdash = combined[combined.TransactionDate.str.contains(pat=r'^\d{1,2}-')].copy()
dfdash = combined[combined.PostedDate.str.contains(pat=r'^\d{1,2}-')].copy()
dfdot = combined[combined.TransactionDate.str.contains(pat=r'^\d{1,2}\.')].copy()
dfdot = combined[combined.PostedDate.str.contains(pat=r'^\d{1,2}\.')].copy()
dfslash = combined[combined.TransactionDate.str.contains(pat=r'^\d{1,2}/')].copy()
dfslash = combined[combined.PostedDate.str.contains(pat=r'^\d{1,2}/')].copy()
dfdash.TransactionDate = pd.to_datetime(dfdash.TransactionDate, format='%d-%b-%y') 
dfdash.PostedDate = pd.to_datetime(dfdash.PostedDate, format='%d-%b-%y') 
dfdot.TransactionDate = dfdot.TransactionDate.str.strip()
dfdot.TransactionDate = pd.to_datetime(dfdot.TransactionDate, format='%d.%m.%Y') 
dfdot.PostedDate = dfdot.PostedDate.str.strip()
dfdot.PostedDate = pd.to_datetime(dfdot.PostedDate, format='%d.%m.%Y') 
dfslash.TransactionDate = dfslash.TransactionDate.str.strip()
dfslash.TransactionDate = pd.to_datetime(dfslash.TransactionDate, format='%m/%d/%Y')
dfslash.PostedDate = dfslash.PostedDate.str.strip()
dfslash.PostedDate = pd.to_datetime(dfslash.PostedDate, format='%m/%d/%Y')
```

```python
dfNew = pd.concat([dfdash,dfdot,dfslash])
dfOSU=dfNew[dfNew['AgencyName'] == 'OKLAHOMA STATE UNIVERSITY']
dfOSU.info()
```

```python
def count_first_digit(data_str):
    mask=dfOSU2014[data_str]>=1
    data=list(dfOSU2014[mask][data_str]) 
    for i in range(len(data)):
        data[i] =str(data[i])
        data[i]=data[i][0]
    first_digits=[int(x) for x in data]
    unique=(set(first_digits))#a list with unique values of first_digit list
    data_count=[]
    for i in unique:
        count=first_digits.count(i)
        data_count.append(count)
    total_count=sum(data_count)
    data_percentage=[(i/total_count)*100 for i in data_count]
    for i in range(len(data_percentage)):
        data_percentage[i] = round(data_percentage[i],1)
    Fraud_test_1 = pd.DataFrame({'Digits':[1,2,3,4,5,6,7,8,9],'Percentage':data_percentage})
    return  Fraud_test_1
   
count_first_digit('Amount')
```
